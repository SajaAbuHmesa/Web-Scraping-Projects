```python
# import required modules
from bs4 import BeautifulSoup
import requests
 
# get URL
page = requests.get("https://en.wikipedia.org/wiki/Main_Page")
 # page object contain 2 elements 1// page.content (the html code structure ) 2// page.status_code (which contain http status code )
# scrape webpage
# display status code as a raw data 
print(page.status_code)
 
# display scraped data
print(page.content)
```

###  Introduction to Beautiful Soup for page parsing

```python
# scrape webpage
soup = BeautifulSoup(page.content, 'html.parser')
 
# display scraped data
print(soup.prettify())
```

### Digging deep into Beautiful Soup further

```python
list(soup.children)
 
# find all occurrence of p in HTML
# includes HTML tags
print(soup.find_all('p'))
 
print('\n\n')
 
# return only text
# does not include HTML tags
print(soup.find_all('p')[0].get_text())
```

###  Exploring page structure with Chrome Dev tools and extracting information

```python
# create object
object = soup.find(id="mp-left")
 
# find tags
items = object.find_all(class_="mp-h2")
result = items[0]
 
# display tags
print(result.prettify())
```
